# 06 Group Aggregation

## Group aggregation

```python
# 기준에 따라 그룹화 -> 함수를 통해 집계 -> 그룹별 집계된 결과 
```

```python
# 데이터프레임 생성; DataFrame()
g = pd.DataFrame({'key': ['A','B','C','A','B','C','A','B','C'],
                  'data': [0,5,10,5,10,15,10,15,20]})
g
```

```python
# 기준에 따라 그룹화; groupby()
grouped = g.groupby('key')['data'] # key값을 기준으로 그룹화 
for name, group in grouped:
    print(name)
    print(group)
```

```python
# 원하는 그룹의 데이터 추출; get_group()
grouped.get_group('A')
```

## 전형적인 사용

```python
# 전형적인 사용(1); key(기준)별 데이터 집계
# 내장함수 직접적으로 사용
```

```python
g.groupby('key')['data'].sum()
```

```python
g.groupby('key')['data'].size()
```

```python
# 전형적인 사용(2); key(기준)별 데이터 집계 + reset_index()사용
# reset_index의 역할
```

```python
g.groupby('key')['data'].sum().reset_index()
```

## Using agg method

```python
# agg(내가 사용하고 싶은 함수)
g.groupby('key')['data'].agg('max')
```

```python
# 여러 개의 집계함수를 사용하고 싶을 때
g.groupby('key')['data'].agg(['max','min'])
```

```python
# 사용자 지정함수를 통해 집계를 하고 싶을 때
g.groupby('key')['data'].agg(lambda x: x.max()-x.min())
```

## Group Aggregation with df

```python
# 1. 특정 기준열에 대해 특정 열의 데이터를 특정 방식으로 집계하고 싶을 때
# Q. 확진일 별로 영등포구에 몇 명의 확진자가 발생했는지 확인하고 싶다
# df를 활용해 다시 Group Aggregation 진행
df.groupby('확진일')['영등포구'].sum().reset_index()
```

```python
# 집계된 데이터의 column명을 바꾸고 싶을 때
# DataFrame.groupby('기준열')['집계할 데이터열'].agg([('집계된 데이터의 새로운 column명', 적용할 함수)])
df.groupby('확진일')['영등포구'].agg([('영등포구일일확진자수',np.sum)]).reset_index()
```

```python
# 2. 특정 기준열에 대해 특정 열의 데이터를 여러 방식으로 집계하고 싶을 때
# Q. 거주구 별로 최초 확진일과 가장 최근 확진자 발생일
# DataFrame.groupby('기준열').agg({'집계할 특정 열': [('첫번째 열의 함수 1로 집계된 새로운 column명', 적용할 함수1),
#                                                     ('첫번째 열의 함수2로 집계된 새로운 column명', 적용할 함수2)]})
df.groupby('거주구').agg({'확진일':[('최초 확진일', np.min),('가장 최근 확진자 발생일', np.max)]}).reset_index()
```

```python
# 3. 특정 기준열에 대해 여러 열의 데이터를 여러 방식으로 집계하고 싶을 때
# Q. 거주구별 최초 확진일과 거주구별 감염경로의 다양성
df.groupby('거주구').agg({'확진일':[('최초 확진일', np.min)],
                      '감염경로':[('감염경로의 다양성', lambda x: x.nunique())]}).reset_index()
```

```python
# 4. 특정 기준열에 대해 여러 열의 데이터를 여러 방식으로 집계하고 싶을 때
# 앞선 방식들을 종합적으로 사용
df.groupby('거주구').agg({'확진일':[('최초 확진일', np.min), ('가장 최근 확진자 발생일', np.max)],
                      '감염경로':[('감염경로의 다양성', lambda x: x.nunique())]}).reset_index()
```

```python
# 5. 여러 기준열에 대해 특정 열의 데이터를 특정 방식으로 집계
# Q. 거주구별로 확진일별로 감염경로의 다양성
# DataFrame.groupby(['기준열1','기준열2'])['집계할 데이터열'].agg([('집계된 데이터의 새로운 column명', 적용할 함수)])
df.groupby(['거주구', '확진일'])['감염경로'].agg([('감염경로의 다양성', lambda x: x.nunique())]).reset_index()
```

# 07 Merging

## Merging DataFrames

```python
# 병합; pd.merge(df1, df2)
# 디폴트; on=None(공통인 모든 열을 키로 인식), how='inner'(공통으로 존재할 때만 추출)
# 옵션; on='특정열', how='outer','inner','left','right'
```

```python
# 데이터프레임 생성

from pandas import DataFrame, Series

df1 = DataFrame({'key': ['b','b','a','c','a','a','b'],
                'data1': range(7)})
df2 = DataFrame({'key': ['a','b','d'], 'data2': range(3)})

display(df1)
display(df2)
```

```python
# how='outer'; 기준열의 데이터가 한 쪽에만 속하더라도 포함
pd.merge(df1, df2, how='outer')
```

```python
# how='inner'; 기준열의 데이터가 양쪽에 존재하는 경우만 포함 
pd.merge(df1, df2, how='inner')
```

```python
# how='left'; 기준열의 데이터를 왼쪽 df 기준으로 병함 
pd.merge(df1, df2, how='left')
```

```python
# how='right'; 기준열의 데이터를 오른쪽 df 기준으로 병함 
pd.merge(df1, df2, how='right')
```
