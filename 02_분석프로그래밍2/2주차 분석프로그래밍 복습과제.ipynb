# 문제3

영화별로 장르의 개수를 확인할 수 있는 코드를 작성해주세요!(함수매핑을 활용하세요!)

```python
def f(x): # 사용자 정의 함수 
    res = 1 # 기본적으로 1개
    for i in x: # 문자열 안에
        if i == '|': # '|' 기호를 만나면 
             res += 1 # +1 해주기 
    return res
```

```python
print(data['genres'].head()) # genres
print(data['genres'].apply(f).head()) # genres의 개수
```
