## 막대 그래프

```python
# 세로형 막대그래프; kind='bar'

data = [["Rudra", 23, 156, 70],
        ["Nayan", 20, 136, 60],
        ["Alok", 15, 100, 35],
        ["Prince", 30, 150, 85]]

df = pd.DataFrame(data, columns=['Name', 'Age', 'Height(cm)', 'Weight(kg)'])

df.plot(x="Name", y=["Age","Height(cm)","Weight(kg)"], kind="bar", figsize=(9,8))
plt.show()
```

```python
# 가로형 막대그래프; kind='barh'

df.plot(x="Name", y=["Age","Height(cm)","Weight(kg)"], kind="barh", figsize=(9,8))
plt.show()
```

```python
# 누적 그래프; stacked=True

employees = ["Rudra", "Alok", "Prince", "Nayan", "Reman"]
earnings = {"January":[10,20,15,18,14],
            "February":[20,13,10,18,15],
            "March":[20,20,10,15,18]}

df = pd.DataFrame(earnings, index=employees)

df.plot(kind="bar", stacked=True, figsize=(10,8))
plt.legend(loc="lower left", bbox_to_anchor=(0.8, 1.0))
plt.show()
```

## 파이차트

```python
# 파이차트1

ratio = [34,32,16,18] # 각 영역의 비율
labels = ['Apple', 'Banana', 'Melon', 'Grapes'] # 각 영역의 이름
explode = [0.05, 0.05, 0.05, 0.05] # 각 영역이 파이차트 중심에서 벗어나는 정도
colors = ['silver', 'gold', 'whitesmoke', 'lightgray'] # 각 영역의 색상

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
# 파이차트2

colors = ['#ff9999','#ffc000','#8fd9b6','#d395d0'] # Hex code를 이용해 더욱 다양한 색상 지정
wedgeprops = {'width':0.7, 'edgecolor':'w', 'linewidth':5} # 부채꼴 영역의 스타일 설정

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
# (1) 축 이름 간결화; xticks, yticks

# tick이 위치할 리스트 
index = np.arange(len(year))

# 이때, year대신 index
plt.plot(index, goals)
plt.title('5년간 연도별 메시의 골 수')
plt.xlabel('연도')
plt.ylabel('골수')

# 설정한 위치에 표시할 labels을 golas로 설정
plt.xticks(index, year) 
plt.show()
```

# lmplot

```python
# lmplot; 회귀선 + 스캐터 플롯
# anscombe; seaborn 내장 데이터셋, 데이터 시각화의 중요성 강조를 위한 데이터

ans = sns.load_dataset('anscombe')
ans.head()
```

```python
# 기본형

sns.lmplot(x='x', y='y', col='dataset', hue='dataset', data=ans,
          ci=False, order=1, robust=False)
```

```python
# cf. regplot
# Q. 왜 lmplot을 주로 쓸까?
# A. col, hue 옵션으로 범주형 변수를 지정하여 카테고리별 데이터 분포의 선형관계 표현 가능

sns.regplot(x='x', y='y', data=ans)
```

```python
# col, hue X

sns.lmplot(x='x', y='y', data=ans,
          ci=False, order=1, robust=False)
```

```python
# hue X

sns.lmplot(x='x', y='y', data=ans, col='dataset',
          ci=False, order=1, robust=False)
```

```python
# ci; 회귀추정의 신뢰구간, 1~100의 값

sns.lmplot(x='x', y='y', data=ans,
          ci=50, order=1, robust=False)
```

```python
# robust; 튀는 값으로 인한 영향 제거 

sns.lmplot(x='x', y='y', data=ans.query("dataset=='III'"), ci=None, robust=True) # robust=False -> True
```

```python
# order; 회귀선의 차원 옵션 변경 가능

sns.lmplot(x='x', y='y', data=ans.query("dataset=='II'"), ci=None, robust=False, order=2) # 2차원 
```

## 여러 개의 plot 그리기

```python
# (1) 겹쳐 그리기
# 최종적으로 plt.show()하기 전 한 셀에 여러 개의 그래프를 넣어줌
# 서로 다른 종류의 그래프도 겹쳐 그릴 수 있음 

sns.distplot(titanic_low.query("sex=='male'")["fare"], hist=False)
sns.distplot(titanic_low.query("sex=='female'")["fare"], hist=False)
plt.show()
```

```python
# cf. for문으로 한 번에 해결

fct = titanic_low['sex'].unique()
for f in fct:
    sns.distplot(titanic_low.query("sex==@f")["fare"], hist=False)

plt.show()
```

```python
# (2) 나란히 배치하기; Subplot
# Subplot; Figure 내에 여러 개의 Axes 객체를 만들어 각각의 Axes에 개별적인 plot 생성
# 양 옆으로 2개의 subplot 만들기

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.boxplot(x='class', y='fare', data=titanic, ax=ax1)
sns.countplot(x='class', data=titanic, ax=ax2)

plt.show()
```

```python
# cf. 더 간단하게; plt.subplots()

fig, ax = plt.subplots(ncols=2)

sns.boxplot(x='class', y='fare', data=titanic, ax=ax[0])
sns.countplot(x='class', data=titanic, ax=ax[1])

plt.show()
```

```python
# (3) 나란히 배치하기 - FacetGrid
# FacetGrid; 범주형 변수의 조합으로 그래프를 나누어 그릴 수 있음
# 한 종류의 그래프를 범주형 변수별로 나누어 볼 때 유리

# 예시; hue를 지정하니 오히려 더 지저분한 그래프
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
```

```python
# 조건에 따라 그리드 나누기

grid = sns.FacetGrid(data=tips, col='day', hue='day', col_wrap=2)

# 나눠진 그리드에 그래프 적용하기

grid.map(sns.scatterplot, "total_bill", "tip")
```

```python
# Subplot vs FacetGrid
# 동일한 종류의 plot을 변수별로 보고싶을 때는 facetGrid가 압도적으로 유리
# ex. outlier를 제거한 Titanic_low에서 성별에 따른 fare 변수의 분포를 확인할 때

# Subplot으로 그리기
fig, ax = plt.subplots(ncols=2)
sns.distplot(titanic_low.query("sex=='male'")['fare'], hist=False, color='steelblue', ax=ax[0])
sns.distplot(titanic_low.query("sex=='female'")['fare'], hist=False, color='darkorange', ax=ax[1])

ax[0].set_title("sex=male")
ax[1].set_title("sex=female")

ax[0].set_ylabel("")
ax[1].set_ylabel("")
ax[1].axes.yaxis.set_visible(False)

plt.show()
```

```python
# FacetGrid로 그리기

grid = sns.FacetGrid(data=titanic_low, col="sex", hue='sex')
grid.map(sns.distplot, 'fare', hist=False)
```
